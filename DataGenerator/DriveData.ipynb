```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def date_parser(string_list):
    return [time.ctime(float(x)) for x in string_list]

drivedata = pd.read_csv('../data/drivedata.csv', parse_dates = [0], date_parser=date_parser)
```

```python
drivedata.query('fuel_injection.isnull()', engine='python')
```

```python
plt.plot(drivedata['time'], drivedata['car_speed'])
```

```python
t = 0
for k, r in drivedata.iterrows():
    tm = int(r['time'])
    if (t > 0) and (t + 1< tm):
        print(str(k) + ':' + str(tm))
    t = tm
```
